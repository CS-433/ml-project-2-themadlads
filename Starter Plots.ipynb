```python
df = df.rename(columns={"Non_rhymes\t": "Non_rhymes"})
```

```python
plt.xticks(rotation=45)
sns.lineplot(df_crpo, x="Name", y="Accuracy", label='Accuracy')
sns.lineplot(df_crpo, x="Name", y="Perfect", label='Perfect rhymes')
sns.lineplot(df_crpo, x="Name", y="Assonant", label='Assonant rhymes')
```
